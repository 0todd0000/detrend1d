# Metadata

The imported NPZ files contain the following metadata variables:

- `cond` : (nsteps,) array of integers specifying condition
- `tsteps` : (nsteps,) array of floats specifying within-session step times
- `sess` : (nsess,) list of session labels
- `tsess` : (nsess,) list of session starting times


````python
import os
import numpy as np
import matplotlib.pyplot as plt
import detrend1d as dtr


subj     = 1
fnameNPZ = os.path.join(dtr.dirDATA, f's{subj:03}.npz')
with np.load(fnameNPZ) as z:
    cond   = z['cond']
    tsteps = z['tsteps']
    sess   = z['sess']
    tsess  = z['tsess']

    
````

This `Metadata` class can be used to conveniently perform various operations on the metadata.

````python
md    = dtr.Metadata(cond, tsess, tsteps, sess)
print( md )
````

Note that the "A" session counters have been dropped from the session labels (i.e., the "A1", "A2", "A3" and "A4" labels are replaced by a single "A" label).

For example, the metadata can be visualized using the `plot` method like this:

````python
plt.figure( figsize=(8,6) )
md.plot()
plt.show()
    
````

<br>
<br>

The `Metadata` class contains the following properties / attributes:

```python
nsess     # [int]  total number of sessions
nsteps    # [int]  total number of steps (i.e, observations) across all sessions

snsteps   # (nsess,) [int]  number of observations by session
scondstr  # (nsess,) [str]  condition string labels by session  (reduced labels)
scond     # (nsess,) [int]  condition integer labels by session (reduced labels)
st        # (nsess,) [float]  session starting times (s)

condstr   # (nsteps,) [str]  condition string labels   (reduced labels)
cond      # (nsteps,) [int]  condition integer labels  (reduced labels)
t         # (nsteps,) [float]  step times (seconds)
tr        # (nsteps,) [float]  step times relative to session start (seconds)
ts        # (nsteps,) [float]  session start times (seconds)
```

<br>
<br>

These can be accessed, for example, as follows:

<br>

````python
print( md.nsess )
print( md.nsteps )
print( md.snsteps )
print( md.scondstr )
````
